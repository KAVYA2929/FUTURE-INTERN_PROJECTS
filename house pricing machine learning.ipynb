```python
categorical_features = train.select_dtypes(include = ["object"]).columns
numerical_features = train.select_dtypes(exclude = ["object"]).columns
numerical_features = numerical_features.drop("SalePrice")
print(str(len(numerical_features)))
print(str(len(categorical_feature)))
train_num = train[numerical_features]
train_cat = train[categorical_features]
```

```python
# Handle remaining missing values for numerical features by using median as replacement by using fillna

print("Before replacing NAs for numerical features in train : " ,(train_num.isnull().values.sum()))
train_num = train_num.fillna(train_num.median())
print("After replacing NAs for numerical features in train : " ,(train_num.isnull().values.sum()))
```

```python
#now calculate skewness
from scipy.stats import skew
skewness = train_num.apply(lambda x:skew(x))
skewness.sort_values(ascending=False)
```

```python
skewness = skewness[abs(skewness)>0.5]
skewness.index #index give the name of subject which have greater than 0.5 skew value
```

```python
skew_features  = train[skewness.index]
skew_features.columns
```

```python
#we can treat skewness of a feature with the help fof log transformation.so we'll apply the same here.
skew_features = np.log1p(skew_features)
```

```python
train_cat = train[categorical_features]
train_cat = pd.get_dummies(train_cat, dtype='int')
train_cat.shape
```

```python
train = pd.concat([train_cat,train_num],axis=1)
train.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(train,y,test_size = 0.3,random_state= 0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
lr
```

```python
prediction = lr.predict(X_test)
```

```python
mse = mean_squared_error(prediction, y_test)
print('mse',mse)
```
